# Q5. Pavaizduokite Baltijos šalių gyventojų skaičių stulpelinėjė diagramoje

```python

bal_countries_list = bal_population.keys()
bal_population_list = bal_population.values()


# Bar Chart
plt.figure(figsize =(14,8))

plt.bar(bal_countries_list, bal_population_list, color="#FF6B6B")

plt.title("\nBaltijos šalių gyventojų skaičius\n", fontsize=22, weight="semibold", color="b")
plt.ylabel("Gyventojų skaičius, mln.\n", fontsize = 18, weight="semibold", color="b")

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)


plt.grid()
plt.show()
```

# Q6. Pavaizduokite Baltijos šalių sostinių gyventojų skaičių stulpelinėjė diagramoje.

```python
bal_countries_c_list = bal_population_c.keys()
bal_population_c_list = bal_population_c.values()

# Bar Chart
plt.figure(figsize =(14,8))

plt.bar(bal_countries_c_list, bal_population_c_list, color="#FF6B6B")

plt.title("\nBaltijos šalių sostinių gyventojų skaičius\n", fontsize=22, weight="semibold", color="b")
plt.ylabel("Gyventojų skaičius, tūkst,\n", fontsize = 18, weight="semibold", color="b")

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)


plt.grid()
plt.show()
```

# Q7. Pavaizduokite Baltijos šalių gyventojų proporcijas skritulinėje diagramoje

```python
plt.figure(figsize =(14,8))

plt.pie(bal_population_list, labels=bal_countries_list, autopct="%1.0f%%", textprops={"fontsize": 14})

plt.title("\nBaltijos gyventojų pasiskirstymas, procentais\n", fontsize=25, weight="bold", color="#247881")

plt.show()
```

# Q8. Suraskite Lietuvos, Latvijos ir Estijos gyventojų skaičiaus dinamiką nuo 2015 iki 2020. Pavaizduokite gyventojų skaičiaus keitimą linijinėje diagramoje.

```python
namedIndex = pd.Series(['2015','2016','2017','2018', '2019','2020'], name = 'Year')
df_baltic_pop = pd.DataFrame({
            'Lithuania': [2.90,4.29,2.83,2.81,2.79,2.79],
            'Latvia': [1.97,1.95,1.94,1.32,1.32,1.32],
            'Estonia': [1.32,1.32,1.32,1.32,1.33,1.33]
            }, index=namedIndex
         )
df_baltic_pop
```

```python
fig, ax = plt.subplots(figsize =(14,8))
df_baltic_pop.plot(ax=ax)

plt.title("\nBalijos šalių gyventojų skaičiaus dinamika, 2015m.-2020m., mln:\n",fontsize=25, weight="bold", color="#247881" )
plt.ylabel("Gyventojų skaičius,2015m.-2020m., mln.\n", fontsize = 18, weight="semibold", color="#247881")
plt.xlabel("Metai", fontsize = 18, color = 'deeppink')

plt.xticks(fontsize=14, color = 'deeppink')
plt.yticks(fontsize=14, color = 'b')

plt.legend(loc="upper right",fontsize=25, bbox_to_anchor=(1.5,1))
plt. grid()
plt.show()
```

# 9. Suraskite Lietuvos, Latvijos ir Estijos gyventojų skaičiaus dinamiką nuo 2015 iki 2020. Pavaizduokite gyventojų skaičiaus dinamiką, panuduojant Grouped Bar Chart tipo grafiką

```python
g7_colors = {
    "Lithuania": "#FF6B6B",
    "Latvia": "#FFD93D",
    "Estonia": "#6BCB77",
}
fig, ax = plt.subplots(figsize=(14,8))
df_baltic_pop.plot.bar(ax = ax, color = g7_colors)

plt.title("\nBaltijos šalių gyventojų skaičiaus dinamika, 2015m.-2020m., mln:\n",fontsize=25, weight="bold", color="#247881" )
plt.ylabel("Gyventojų skaičius, mln.\n", fontsize = 18, weight="semibold", color="#247881")
plt.xlabel("Metai", fontsize = 18, color = 'deeppink')

plt.xticks(fontsize=14, color = 'deeppink')
plt.yticks(fontsize=14, color = 'b')
plt. grid()
plt.legend(loc="upper right",fontsize=25, bbox_to_anchor=(1.4,1))
```
